```python
import os
import random
import pandas as pd

# Folder with your merged episodes
folder = "/mnt/newssd/lerobot/lerobot/datasets/Task_PicupBrick_50fps_with_mountingpoints_cotrain/data/chunk-000"

# How many files to check
num_files_to_check = 5

# List all parquet files in the folder
files = [f for f in os.listdir(folder) if f.endswith('.parquet')]
if not files:
    print("No parquet files found in the folder.")
    exit()

# Randomly select files
selected_files = random.sample(files, min(num_files_to_check, len(files)))

for fname in selected_files:
    fpath = os.path.join(folder, fname)
    df = pd.read_parquet(fpath)
    # Pick a random row in the dataframe
    idx = random.randint(0, len(df) - 1)
    obs = df['observation.state'].iloc[idx]
    act = df['action'].iloc[idx]
    print(f"\nFile: {fname}")
    print(f"  Random row index: {idx}")
    print(f"  observation.state length: {len(obs)}")
    print(f"  action length: {len(act)}")
    print(f"  observation.state (first 5): {obs[:5]}")
    print(f"  action (first 5): {act[:5]}")
```
